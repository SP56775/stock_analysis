```python
#Understand the historical performance, return behavior, and volatility of a stock.
#We'll analyze:
#- Price trends over time
#- Daily returns
#- Volatility patterns
#- Moving averages to identify trends
```

```python
#Importing the libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Collect the Data
```

```python
stock=yf.download('RELIANCE.NS',
start='2020-01-01',
end='2024-12-24')
stock.head()
```

```python
#Plot Closing Price
stock['Close'].plot(figsize=(12,6),title='Reliance_closing_price')
plt.savefig('closing_price.png')
```

```python
#Calculation returns
stock['Daily Return']=stock['Close'].pct_change()
stock['Daily Return'].plot(figsize=(15,15),title='Reliance_daily_return')
plt.savefig('Reliance_daily_return.png')
```

```python
#Calculate Moving Averages (SMA30, SMA50)
stock['SMA30']=stock['Close'].rolling(window=30).mean()
stock['SMA50']=stock['Close'].rolling(window=50).mean()
```

```python
stock['SMA30']
```

```python
stock['SMA50']
```

```python
stock[['Close','SMA30','SMA50']].plot(figsize=(15,15),title='moving average')
plt.savefig('moving average.png')
```

```python
#Plot Volatility 
stock['Volatitlity']=stock['Daily Return'].rolling(window=30).std()
stock['Volatitlity'].plot(figsize=(15,15),title='30-Day Rolling Volatility')
plt.savefig('Volatility.png')
```
